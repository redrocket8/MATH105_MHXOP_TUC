#MATH 105 Theory Matplotlib

```python
#pip install matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
decades = ['1980-1990', '1990-2000', '2000-2010','2010-2020']
earthquakes_Crete = [100, 40, 80, 110]
#earthquakes_Atikh = [60, 80, 70, 65]
style.use('ggplot')
plt.figure(figsize=(10,5))
plt.title("Earthquakes per decade")
plt.bar(decades, earthquakes_Crete, color='green', width=0.5, label='Crete')
#plt.bar(decades, earthquakes_Atikh, color='blue', label='Atikh', width =0.5)
plt.xlabel('decades')
plt.ylabel('#earthquakes')
plt.legend()
plt.savefig('test.png')
plt.show
```

```python
decades = ['1980-1990', '1990-2000', '2000-2010','2010-2020']
earthquakes_Crete = [100, 40, 80, 110]
earthquakes_Atikh = [60, 80, 70, 65]
```

```python
xpos=np.array([0,1,2,3])#np.arange(len(decades))
style.use('ggplot')
plt.figure(figsize=(6,5))
barwidth = 0.2
plt.bar(xpos,earthquakes_Crete, color = 'green', width=barwidth, label= 'Crete')
plt.bar(xpos+0.2, earthquakes_Atikh, color = 'red', width=barwidth, label = 'Atikh')

plt.title('Earthquakes per decade \n Crete vs Atikh')
plt.xlabel('decades')
plt.ylabel('# earthquakes')
plt.legend()
plt.xticks(xpos+0.1, decades)
```

```python
ypos=np.arange(len(decades))
style.use('ggplot')
plt.figure(figsize=(10,5))
decades = ['1980-1990', '1990-2000', '2000-2010','2010-2020']
earthquakes_Crete = [100, 40, 80, 110]
earthquakes_Atikh = [60, 80, 70, 65]
barwidth=0.2
plt.barh(ypos,earthquakes_Crete, color = 'green', height=barwidth, label= 'Crete')
plt.barh(ypos+0.2,earthquakes_Atikh, color='red',height=barwidth, label='Atikh')
plt.yticks(ypos+0.1,decades)
plt.xlabel('# earth quakes')
plt.ylabel('decades')
plt.title(' Number of earthquakes per decade\n Crete vs Atikh')
plt.show()
```
