## **Εισαγωγή**

Η δημιουργία και η προσπέλαση αρχείων είναι κρίσιμες λειτουργίες στον προγραμματισμό, που επιτρέπουν στους προγραμματιστές να αποθηκεύουν και να ανακτούν δεδομένα από εξωτερικά αρχεία. Στη γλώσσα προγραμματισμού Python, η διαχείριση αρχείων γίνεται με τη χρήση της συνάρτησης `open()`.

Κατά το άνοιγμα ενός **αρχείου**, προσδιορίζουμε το όνομά του και τον τρόπο ανοίγματος, που καθορίζει εάν θα διαβάζουμε ('r'), θα γράφουμε ('w'), ή θα προσθέτουμε περιεχόμενο ('a'). Είναι σημαντικό να κλείνουμε το αρχείο μετά τη χρήση, χρησιμοποιώντας τη μέθοδο `close()`.

Για την ανάγνωση των περιεχομένων ενός αρχείου, χρησιμοποιούμε μεθόδους όπως `read()`, `readline()` ή `readlines()`. Η εγγραφή σε ένα αρχείο γίνεται με τη μέθοδο `write()`.

Στην Python, η διαχείριση αρχείων μας παρέχει την ευελιξία να διαχειριζόμαστε δεδομένα σε μια ποικιλία μορφών, πετυχαίνοντας την αποτελεσματική αποθήκευση και ανάκτηση πληροφοριών από αρχεία.

## **Άνοιγμα αρχείου**

```python
open('filename', 'mode')
```
`'filename'`: Το όνομα του αρχείου που θέλουμε να ανοίξουμε.

`'mode'`: Καθορίζει τον τρόπο προσπέλασης του αρχείου.


*   `'filename'`: Το όνομα του αρχείου που θέλουμε να ανοίξουμε.
*   `'mode'`: Καθορίζει τον τρόπο προσπέλασης του αρχείου.
    * `'r'` Ανοίγει ένα αρχείο κειμένου για ανάγνωση (read),
    * `'w'` Ανοίγει ένα αρχείο κειμένου για γράψιμο (write). Αν το αρχείο υπάρχει τα περιεχομενά του διαγράφονται.
    * `'a'` Ανοίγει ένα αρχείο κειμένου για προσθήκη (append), δηλαδή προσθέτει τα νέα δεδομένα στο τέλος του αρχείου. Αν το αρχείο δεν υπάρχει το δημιουργεί.

### $\color{green}{\text{Παράδειγμα 4ο}}$:
Ο παρακάτω κώδικας ανοίγει το αρχείο **example.txt** σε read mode, διαβάζει τις γραμμές του αρχείου και τις αποθηκεύει σε μία λίστα

````python
#code_64a.py

file = open('example.txt', 'r')
content = file.readlines() # read all the lines of the file and save them to a list
print(content)

file.close()
````

Εναλλακτικά, μπορούμε να εφαρμόσουμε και τον παρακάτω τρόπο για να διαβάσουμε το αρχείο

````python
#code_64b.py

content=[] # creat an empty list

with open('example.txt', 'r') as file: # open the file in read ('r') mode
    for line in file:                  # itterate the lines of the file
        content.append(line)           # save each line into the list

print(content)
````

Οι κύριες διαφορές των δύο τρόπων είναι:

1.   **`readlines()` method**:
 *   Διαβάζει όλες τις γραμμές του αρχείου και επιστρέφει μία λίστα όπου το κάθε στοιχείο της είναι μία γραμμή από το αρχείο.
 *   Όλο το περιεχόμενο του αρχείου φορτώνεται στην μνήμη με τη μία, το οποίο μπορεί να μην είναι memory-efficient για μεγάλα αρχεία.
2.   **`for line in file`**:
 *  Διαβάζει το αρχείο γραμμή κατα γραμμή δίχως να φορτώνει όλο το περιεχόμενο του αρχείου στην μνήμη.
 * Είναι περισσότερο memory-efficient, κυρίως, για μεγάλα αρχεία αφού μόνο μία γραμμή την φορά αποθηκεύεται στην μνήμη.

### $\color{green}{\text{Παράδειγμα 5ο}}$:
Ο παρακάτω κώδικας δημιουργεί ένα αρχείο με το όνομα **numbers.dat** και γράφει σε αυτο 50 φυσικούς αριθμούς καθώς και τα τετραγωνά τους σε δύο στήλες.

````python
#code_65.py

#open a new file 'numbers.dat' in write mode and write in each line i  i^2
output_file_name='numbers.dat'

with open(output_file_name,'w') as output_file: #
    for i in range(1,101):

        output_file.write(f"{i}\t{i**2}\n")

        print(f"{i}\t{i**2}\n")
````

### $\color{green}{\text{Παράδειγμα 6ο}}$:
Έστω ότι θέλουμε να προσθέσουμε στο αρχείο που δημιουργήσαμε στο *Παράδειγμα 5* στο τέλος του αρχείου, τον μέσο όρο της πρώτης στήλης και της δεύτερης.
*   Αρχικά θα ανοίξουμε το αρχείο **numbers.dat** σε *read mode* και θα αποθηκεύσουμε τα στοιχεία της κάθε στήλης σε δύο λίστες **column1** and **column2** των οποίων θα υπολογίσουμε τον μέσο όρο των στοιχείων τους.

````python
#code_66a.py
column1=[]
column2=[]
with open('numbers.dat','r') as input_file:
    for line in input_file:
        columns=line.strip().split()  # what does strip and split do?
        column1.append(float(columns[0]))
        column2.append(float(columns[1]))

sum1=0.0
sum2=0.0
for i in range(len(column1)):
    sum1=sum1+column1[i]
    sum2=sum2+column2[i]
````

*     Στην συνέχεια θα ανοίξουμε το αρχείο **numbers.dat** σε *append mode*

````python
#code_67b.py
with open('numbers.dat','a') as output_file:
    output_file.write(f"Ο μέσος όρος των στοιχείων της πρώτης στήλης είναι {sum1/len(column1)}.\nΟ μέσος όρος της δεύτερης στήλης είναι {sum2/len(column2)}.\n")
````
